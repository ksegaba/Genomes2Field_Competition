## Levels of missing data in the training data files

```python
import os
import datatable as dt
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
os.chdir('/mnt/home/seguraab/Shiu_Lab/Collabs/Maize_GxE_Competition_Data/Training_Data')
def get_missing(f):
    print(f)
    df = dt.fread(f)
    df = df.to_pandas()
    # column-wise
    print("Column-wise")
    print(df.isna().sum(axis=0)/df.shape[0])
    # row-wise
    print("Row-wise")
    print(df.isna().sum(axis=1)/df.shape[1])
    return (df)

f = '1_Training_Trait_Data_2014_2021.csv'
df = get_missing(f)
ax = df.hist(figsize=(15,15), align='mid')
plt.savefig(f+'.png')

f = '2_Training_Meta_Data_2014_2021.csv'
df = get_missing(f)
ax = df.hist(figsize=(15,15), align='mid')
plt.savefig(f+'.png')

f = '3_Training_Soil_Data_2015_2021.csv'
df = get_missing(f)
ax = df.hist(figsize=(15,15), align='mid')
plt.savefig(f+'.png')

f = '4_Training_Weather_Data_2014_2021.csv'
df = get_missing(f)
ax = df.hist(figsize=(15,15), align='mid')
plt.savefig(f+'.png')

f = '6_Training_EC_Data_2014_2021.csv'
df = get_missing(f)
```
